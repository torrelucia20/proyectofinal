```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#Dataset AIRBNB
#Carga de Dataset utilizando funciones de Pandas
import pandas as pd
from google.colab import drive
import os
drive.mount('/content/gdrive')

%cd '/content/gdrive/MyDrive/DATA_SCIENCE_LUCIA_TORRE/ARCHIVOS_CSV'
airbnb_dataset= pd.read_csv('Airbnb.csv',sep=',')
```

```python
grouped_by_neighbourhood = airbnb_dataset.groupby("neighbourhood group")
grouped_by_neighbourhood_counted = grouped_by_neighbourhood.count()

fig,ax = plt.subplots(figsize =(10, 7))
ax.pie(grouped_by_neighbourhood_counted.id, labels = grouped_by_neighbourhood.groups.keys())

#Este gráfico representa la cantidad de hospedajes que hay por cada grupo de barrio
```

```python
grouped_by_neighbourhood_sum = grouped_by_neighbourhood.sum()
#print (grouped_by_neighbourhood_sum["number of reviews"])

fig, ax = plt.subplots(figsize=(10,8))
ax.bar(grouped_by_neighbourhood.groups.keys(), grouped_by_neighbourhood_sum["number of reviews"])
ax.set_title('Cantidad de opiniones por los usuarios segun el barrio')
ax.set_ylabel('Cantidad de opiniones registradas (millones)')
ax.set_xlabel('Barrio')

#Este gráfico de barras representa la cantidad de opiniones (reviews) que hay por grupo de barrio
```

```python
#airbnb_dataset.values.flatten()

fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(airbnb_dataset["minimum nights"], bins=10)
ax.set_title('Minimo de noches')
ax.set_xlabel('Minimo de noches')
ax.set_ylabel('Frecuencia')

#grouped_by_neighbourhood_mean = grouped_by_neighbourhood.mean()
#print(grouped_by_neighbourhood_mean)

#Lo que se grafica en este histograma es la variable elegida (minimum nights) y su frecuencia respecto a la lista total de datos.
#Se tienen una gran variedad de datos (dato máximo=5645), pero los más frecuentes (que se repiten más en el dataset) se observa que están entre 0 y alrededor de 80. 

```

```python
grouped_by_host_identity_verified = airbnb_dataset.groupby("host_identity_verified")
grouped_by_host_identity_verified_counted = grouped_by_host_identity_verified.count()

fig,ax = plt.subplots(figsize =(10, 7))
ax.pie(grouped_by_host_identity_verified_counted.id, labels = grouped_by_host_identity_verified.groups.keys())

#Este gráfico representa la cantidad de hospedadores verificados y no verificados.
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")
ax = sns.barplot(x="neighbourhood group", y="number of reviews", data=airbnb_dataset)

#Este gráfico intentó representar nuevamente la cantidad de evaluaciones (reviews) con respecto a los distintos grupos de barrio
```

```python
plt.figure()
df_cy_2022 = airbnb_dataset.query("`Construction year`==2022")
ax = sns.displot(data=df_cy_2022, x='review rate number', kind='kde', hue='neighbourhood group', fill=True)
ax.set(xlabel='Review rate', ylabel='Densidad', title='Distribución  de las Reviews')

# En este gráfico se representa las diferentes calificaciones con las que fueron puntuados los distintos grupos de barrios. (Review rate number)
#También pueden verse cuales fueron los barrios que recibieron más puntuaciones
```

```python
plt.figure()
ax = sns.displot(data=airbnb_dataset, x='review rate number', kind='hist')
ax.set(xlabel='Review rate', ylabel='Densidad', title='Distribución  de las Reviews')

#En este gráfico se pueden observar las distintas puntuaciones realizadas por los usuarios
#Puede observarse que la minoría de las evaluaciones obtuvieron la calificación de 1
```
