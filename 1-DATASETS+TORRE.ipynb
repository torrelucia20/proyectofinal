# Nueva sección

Primer database: BANK


Un gerente del banco está molesto porque cada vez más clientes abandonan sus servicios de tarjeta de crédito. El objetivo de este trabajo es poder predecir que usuarios van a querer abandonar, para que el banco puedan ir proactivamente al cliente a brindarle mejores servicios, y hacer que continúen siendo sus clientes.


Diccionario de datos

- CLIENTNUM: Número de cliente. Identificador único del cliente titular de la cuenta
- Attrition_Flag: actividad del cliente, si la cuenta está cerrada, entonces 1 sino 0
- Customer_Age: Variable demográfica - Edad del cliente en años
- Género: Variable demográfica - M=Hombre, F=Mujer
- Dependent_count: Variable demográfica, Número de dependientes
- Education_Level: Variable demográfica, calificación educativa del titular de la cuenta (ejemplo: escuela secundaria, graduado universitario, etc.)
- Marital_Status: Variable demográfica, Casado, Soltero, Divorciado, Desconocido
- Categoría de ingreso Variable demográfica: Categoría de ingreso anual del titular de la cuenta (< $40K, $40K - 60K, $60K - $80K, $80K-$120K, >
- Card_Category: Producto Variable, Tipo de Tarjeta (Azul, Plata, Oro, Platino)
- Monthsonbook: Período de relación con el banco
- TotalRelationshipcount: Número total de productos en poder del cliente
- MonthsInactive12_mon: Número de meses inactivos en los últimos 12 meses
- ContactsCount12_mon: Número de contactos en los últimos 12 meses
- Credit_Limit: Límite de crédito en la tarjeta de crédito
- TotalRevolvingBal: Saldo total rotativo en la tarjeta de crédito
- AvgOpenTo_Buy: Línea de crédito abierta para comprar (promedio de los últimos 12 meses)
- TotalAmtChngQ4Q1: Cambio en el monto de la transacción (Q4 sobre Q1)
- TotalTransAmt: Importe total de la transacción (últimos 12 meses)
- TotalTransCt: Recuento total de transacciones (últimos 12 meses)
- TotalCtChngQ4Q1: Cambio en el recuento de transacciones (Q4 sobre Q1)
- AvgUtilizationRatio: Promedio de uso de tarjeta
- NaiveBayesClassifier_attribution: Naive Bayes

Para predecir que características tienen los usuarios que tienen mayor probabilidad de abandonar los servicios de sus tarjetas de crédito pueden ser útiles los siguientes grupos de variables:
- Información demográfica del usuario (Edad, genero, educación, estado marital, sueldo)
- Información tipo de tarjeta de crédito (límite de credito, tiempo registrado con el banco)
- Información del uso de la tarjeta de crédito (total de transacciones, promedio de uso de tarjeta.)

- lat: latitud
- long: longitud
- country: país
- country code: código de país
- instant_bookable: Se puede reservar instantáneamente

- service fee: tarifa de servicio

- number of reviews: numéro de evaluaciones realizadas
- last review: última evaluación
- reviews per month: evaluaciones por mes
- review rate number: puntuación dentro de las evaluaciones realizadas
- calculated host listings count: recuento calculado de listas de hospedadores
- availability 365: disponibilidad dentro de los 365 días del año
